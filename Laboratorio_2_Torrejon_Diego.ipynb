```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar imagen
from google.colab import files
files.upload()
imagen = cv2.imread('Miles.jpg')

# Convertir de BGR a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar imagen
plt.imshow(imagen_rgb)
plt.title("Imagen Original")
plt.axis("off")
plt.show()
```

```python
# Convertir a escala de grises
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Mostrar imagen en grises
plt.imshow(gris, cmap='gray')
plt.title("Escala de Grises")
plt.axis("off")
plt.show()
```

```python
# Aumentar brillo
brillo = cv2.convertScaleAbs(gris, alpha=1.2, beta=30)

# Mostrar imagen con brillo
plt.imshow(brillo, cmap='gray')
plt.title("Imagen con Brillo alto")
plt.axis("off")
plt.show()
```

```python
brillo = cv2.convertScaleAbs(gris, alpha=2.0, beta=60)
# Mostrar imagen con brillo
plt.imshow(brillo, cmap='gray')
plt.title("Imagen")
plt.axis("off")
plt.show()
```

```python
brillo = cv2.convertScaleAbs(gris, alpha=4.0, beta=80)

plt.imshow(brillo, cmap='gray')
plt.title("Variando")
plt.axis("off")
plt.show()
```

```python
brillo = cv2.convertScaleAbs(gris, alpha=4.0, beta=80)

plt.imshow(brillo, cmap='gray')
plt.title("alpha=4.0 beta=80")
plt.axis("off")
plt.show()
```

```python
bordes = cv2.Canny(gris, 50, 100)

plt.imshow(bordes, cmap='gray')
plt.title("Bordes (100,200)")
plt.axis("off")
plt.show()
```
